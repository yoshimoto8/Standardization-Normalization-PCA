```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#ランダムデータ
data = np.random.randint(0, 20,100)
data2 = np.random.randint(0, 20,100)
random_data = np.array([data, data2]).reshape(100,2)
```

```python
# 線形データ
liner_data = np.linspace(0, 20, 100)
liner_data2 = np.linspace(0, 20, 100)
liner_data3 = np.array([liner_data, liner_data2]).reshape(100,2)
```

```python
def no_processing_scatter(name, scatter_data, scatter_data2):
    plt.xlim([0,20])
    plt.ylim([0,20])
    plt.title(name, fontsize=25, fontname='serif')
    plt.scatter(scatter_data, scatter_data2)

def processing_scatter(name, scatter_data, scatter_data2):
    plt.xlim([-1,1])
    plt.ylim([-1,1])
    plt.title(name, fontsize=25, fontname='serif')
    plt.scatter(scatter_data, scatter_data2)
```

```python
name = "randam data scatter"
no_processing_scatter(name,data, data2)
```

```python
name = "liner data scatter"
no_processing_scatter(name ,liner_data, liner_data2)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
 
# 標準化
sc = StandardScaler()
ramdom_sc = sc.fit_transform(random_data)
sc_data = ramdom_sc[:, :1]
sc_data2 =ramdom_sc[:, 1:2]

name = 'Standardizated random data scatter'
processing_scatter(name,sc_data, sc_data2)
```

```python
liner_sc = sc.fit_transform(liner_data3)
sc_data3 = liner_sc[:, :1]
sc_data4 =liner_sc[:, 1:2]
name = "Standardizated liner data scatter"
processing_scatter(name, sc_data3, sc_data4)
```

```python
# 正規化
ms = MinMaxScaler()
ramdom_ms = ms.fit_transform(random_data)
ms_data = ramdom_ms[:, :1]
ms_data2 =ramdom_ms[:, 1:2]

name = "Normalized  random data scatter"
processing_scatter(name, ms_data, ms_data2)
```

```python
li_ms = ms.fit_transform(liner_data3)
ms_data3 = li_ms[:, :1]
ms_data4 =li_ms[:, 1:2]
name = "Normalized  liner data scatter"
processing_scatter(name,ms_data3, ms_data4)
```

```python
from sklearn.decomposition import PCA
```

```python
#randomPCA
pca = PCA()
pca.fit(random_data)
transformed = pca.fit_transform(random_data)

ra_transformed = transformed[:, :1]
ra_transformed2 = transformed[:, 1:2]
name = "ramdom data PCA"
no_processing_scatter(name, ra_transformed, ra_transformed2)
```

```python
sted = np.array([sc_data, sc_data2]).reshape(100,2)
pca.fit(sted)
st_transformed = pca.fit_transform(sted)
name = "random data PAC after Standardizing"
no_processing_scatter(name,st_transformed[:, :1],st_transformed[:, 1:2])
```

```python
sted = np.array([ms_data, ms_data2]).reshape(100,2)
pca.fit(sted)
ms_transformed = pca.fit_transform(sted)
name = "random data PAC after Normalizing"
no_processing_scatter(name,ms_transformed[:, :1],ms_transformed[:, 1:2])
```

```python
#LinerPCA
pca = PCA()
pca.fit(random_data)
transformed = pca.fit_transform(liner_data3)

ra_transformed = transformed[:, :1]
ra_transformed2 = transformed[:, 1:2]
name = "Liner data PCA"
no_processing_scatter(name, ra_transformed, ra_transformed2)
```

```python
sted = np.array([sc_data3, sc_data4]).reshape(100,2)
pca.fit(sted)
st_transformed = pca.fit_transform(sted)
name = "Liner data PAC after Standardizing"
no_processing_scatter(name,st_transformed[:, :1],st_transformed[:, 1:2])
```
